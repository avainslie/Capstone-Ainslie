## Functions for feature engineering

<b> Rational:</b> Removing special characters, leaving the V's and E's in the diagnosis columns because this is not a mistake, but rather gives information to which diagnosis is being indicated as the letters are part of the code. Later these columns are mapped to numerical values through encoding. Removed values with less than 3 characters from the diagnosis columns as these are most likely typos. Replaced PhysicianNotFound with a NaN since this is more accurate and can be later imputed. Then for all floats and object dtypes removed special characters and replaced with NaNs. Resulting data frame is a mix of string and floats in some columns which is dealt with later in the notebook. Also leaving > character because it is important to the meaning of the values it is in.

````python
# To find all special characters going to check unique values for each col

# for col in train_set.columns:
#     print(col, train_set[col].unique())
````

````python
def remove_special_char(data):
    # remove values with less than 3 digits/char since I think they are typos
    diagnosis=["diag_1","diag_2","diag_3"]
    for col in diagnosis:
        data[col] = data[col].where(data[col].str.len() >= 3) 
        
    data['medical_specialty'] = data.medical_specialty.replace('PhysicianNotFound', np.NaN)
#     data['payer_code']= data.payer_code.replace("?", np.NaN)
#     data['medical_specialty'] = data.medical_specialty.replace("?", np.NaN)
        
    for col in list(data.columns):
        if data[col].dtype == float:
            data[col] = data[col].str.strip().replace('?', np.NaN).astype(float)
            data[col]= data[col].str.strip().replace("\t", np.NaN).astype(float)
            data[col] = data[col].str.strip().replace('.', np.NaN).astype(float)
            data[col] = data[col].str.strip().replace('/', np.NaN).astype(float)
            data[col] = data[col].str.strip().replace('nan', np.NaN).astype(float)
        elif data[col].dtype == object:
#             data[col]= data[col].str.strip().replace("\t", np.NaN)
            data[col] = data[col].str.strip().replace('?', np.NaN).astype(str)
#             data[col] = data[col].str.strip().replace('.', np.NaN)
#             data[col] = data[col].str.strip().replace('/', np.NaN)
#             data[col] = data[col].str.strip().replace('nan', np.NaN)
    return(data)
````

Rational: Used SimpleImputer with the strategies most_frequent and median for binary and continuous variables respectively. This is because using these strategies for these column types gives more accurate values to replace NaNs with. I have explained in more detail in past assignments, but essentially it is more likely that the most_frequent value of a binary column is more representative of the trends of the data than the most_frequent value of a non binary column. For the purpose of this data set I counted columns with 4 or less unique values as 'binary' because these columns were categorical and I am making the assumption that the most frequent value in these columns is also representative of the overall trend of the dataset. This is because since there is only 4 or less values, there is a low likelihood that the most_frequent value is just there because it was repeated one or two times more than the other values. This can be verified by checking df.column.value_counts() to see the amounts of each value per column. 

````python
from sklearn.impute import SimpleImputer

def impute(data):
    bin_imputer= SimpleImputer(strategy="most_frequent")
    everything_else_imputer = SimpleImputer(strategy="median")
    
    data_bin=[]
    data_nonbin=[]
    
    for col in list(data.columns):
        value_counts= list(data[col].value_counts())
        # check if binary, then append col to data_bin
        if len(value_counts) <= 4 or data[col].dtype == object: 
            data_bin.append(col)
        # check if anything else, then append col to data_nonbin
        else:
            data_nonbin.append(col)

    data_bin= data[data_bin].copy()
    data_nonbin= data[data_nonbin].copy()
    
    binary_columns = pd.DataFrame(data_bin,columns=data_bin.columns,index=data_bin.index)
    non_bin_columns= pd.DataFrame(data_nonbin, columns=data_nonbin.columns,index=data_nonbin.index)
    
    #fit_by_nonbin = pd.Series([i for i in non_bin_columns.iloc[:,0].unique() if type(i) == str])
    
    binary_columns= pd.DataFrame(bin_imputer.fit_transform(binary_columns), columns= binary_columns.columns, index= binary_columns.index)
    non_bin_colums = pd.DataFrame(everything_else_imputer.fit_transform(non_bin_columns), columns= non_bin_columns.columns, index= non_bin_columns.index)
    
    non_bin_colums = non_bin_colums.astype(float) # helps prevent rounding errors
    data = pd.concat([binary_columns, non_bin_colums], axis=1) 
    #data = data.astype(float)
    #print(data.columns.isnull().sum())
    return(data)
````

<b> Rational: </b> I used RobustScaler because it is not as sensitive to outliers as other scalers, such as MinMax and Standard. RobustScaler also, unlike other scalers, does not assume a normal distribution in the data. If a normal data distribution is assumed and not true, then the results of scaling will not be as accurate. RobustScaler removes the median and uses the inter quartile range to find the scaled value for each feature. Doing so allows it to be robust to the outliers in the data and to preserve them.

````python
# Taken from lecture 5
def remove_corr(df, threshold):
    
    corr_mx = df.corr() #Correlation matrix
    
    newColumns = [df.columns[0]]   #First column

    for colx in df.columns[1:]:   #Second column on

        if (np.abs(corr_mx.loc[colx, newColumns]) < threshold).all():

            newColumns.append(colx)

        else:

            print('Removing correlated column %s' % colx)
        
            df = df.drop(colx, axis=1)     
            
    return df
````

### Comparision of the base vs modified models

Compare metrics and methods used. 

----

The base model was actually the best model I was able to achieve. This was unexpected. I thought that the under or over sampling would definitely improve the model since it would have a more balanced data set to be trained on this way. I also had a lot of hope in the LinearSVM model since through its hyper parameter tunning I thought it could have found the best way to penalize miss-classifications of the minority class and thus allow the model to better learn the trends of the data.    
XGBoost and gradient boosting both gave low AUC scores of around 0.5. I did not perform any tunning on them since I thought SVM and adaboost had more potential.   

The adaboost algorithm also had hope, I found when I used it with no classifier passed as a parameter, it would return a AUC of 0.61-0.62 when trained on under sampled data. So my thinking was that given a good classifier such as RandomForest, adaboost would be able to improve on this score. This unfortunately was not true and after adding RandomForest, adaboost scored worse. With the addtion of cross validation, I was able to bring the score back up to 0.62. In hopes that this model could be improved I tried randomized search to tune the hyper parameters. With no randomforest classifier the AUC on the train was 0.67. With RandomForest added in, the score was worse at 0.62. Then I tried using the best parameters in the rf classifier thinking this could improve the result. This cause the kernel to be stuck again. Adaboost genrally was a very fast algorithm until this modification. I tried removing the n_estimators from the rf classifier because I thought this is why it was getting stuck, the results were that the kernel still gets stuck. The model that produced a training AUC score of 0.67 only had a score of 0.62 on the test data. I tried multiple ways of fitting the adaboost and none could result in a higher AUC score. 

My last hope was in the VotingClassifier. I was thinking that maybe by combining some models, this classifier would be able to take the results and find a better score. This may not have worked since the individual scores from all the classifiers were on the lower side to begin with, potentially they were all miss-classifying similar results which led to bad "votes" and thus bad results from the ensemble model.   

Right before the deadline I found out there is a parameter for rf that is equivilant to class_weight in SVM. If given more time I would have tried to adjust this hyperparameter to handle the imbalanced data better.
